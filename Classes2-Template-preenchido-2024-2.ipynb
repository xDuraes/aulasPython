## Sobrecarga de Operadores

### Operadores Numéricos

| Operador         | Expressão  | Método Especial            |
| ---------------- | ---------- | -------------------------- |
| Adição           | Q1 + Q2    | \_\_add\_\_(self, Q2)      |
| Subtração        | Q1 - Q2    | \_\_sub\_\_(self, Q2)      |
| Multiplicação    | Q1 \* Q2   | \_\_mul\_\_(self, Q2)      |
| Potência         | Q1 \*\* Q2 | \_\_pow\_\_(self, Q2)      |
| Divisão          | Q1 / Q2    | \_\_truediv\_\_(self, Q2)  |
| Divisão Inteira  | Q1 // Q2   | \_\_floordiv\_\_(self, Q2) |
| Resto da Divisão | Q1 % Q2    | \_\_mod\_\_(self, Q2)      |
| \-               | \-Q1       | \_\_neg\_\_()              |
| +                | +Q1        | \_\_pos\_\_()              |

```python
class Quantidade:
    def __init__(self, val):
        self.__valor = val
        
    def __add__(self, outro):
        print("-------------")
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor + outro
        else:
            novo = self.__valor + outro.__valor
        return Quantidade(novo)
    def __radd__(self, outro):
        print("----R----")
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor + outro
        else:
            novo = self.__valor + outro.__valor
        return Quantidade(novo)
    def __str__(self):
        return "Quantidade(" + str(self.__valor) +")"
    
    
Q1 = Quantidade(10)
Q2 = Quantidade(5)

print(Q1+Q2)
# não altera o valor
Q1.__valor=20
print(Q1+20)
print(Q1+20.98)
print(20.98+Q1)
R1=20.98+Q1
print(type(R1))
```

### Operadores de Comparação

| Operador             | Expressão | Método Especial    |
| -------------------- | --------- | ------------------ |
| Igual                | Q1 == Q2  | \_\_eq\_\_(Q1, Q2) |
| Diferente            | Q1 != Q2  | \_\_ne\_\_(Q1, Q2) |
| Menor que            | Q1 < Q2   | \_\_lt\_\_(Q1, Q2) |
| Menor que ou igual a | Q1 <= Q2  | \_\_le\_\_(Q1, Q2) |
| Maior que            | Q1 > Q2   | \_\_gt\_\_(Q1, Q2) |
| Maior que ou igual a | Q1 >= Q2  | \_\_ge\_\_(Q1, Q2) |

```python
class Quantidade:
    def __init__(self, val):
        self.__valor = val
        
    def __add__(self, outro):
        print("-------------")
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor + outro
        else:
            novo = self.__valor + outro.__valor
        return Quantidade(novo)
    def __radd__(self, outro):
        print("----R----")
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor + outro
        else:
            novo = self.__valor + outro.__valor
        return Quantidade(novo)
    
    def __gt__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            return self.__valor > outro
        else:
            return self.__valor>outro.__valor
    
    
    def __str__(self):
        return "Quantidade(" + str(self.__valor) +")"
    
    
Q1 = Quantidade(10)
Q2 = Quantidade(5)

print(Q1+Q2)
# não altera o valor
Q1.__valor=20
print(Q1+20)
print(Q1+20.98)
print(20.98+Q1)
R1=20.98+Q1
print(type(R1))

print(R1>10)
```

### Operadores com atribuição

| Operador com atribuição  | Expressão   | Método Especial             |
| ------------------------ | ----------- | --------------------------- |
| Adição                   | Q1 += Q2    | \_\_iadd\_\_(self, Q2)      |

| Multiplicação            | Q1 \*= Q2   | \_\_imul\_\_(self, Q2)      |
| Potência                 | Q1 \*\*= Q2 | \_\_ipow\_\_(self, Q2)      |
| Divisão                  | Q1 /= Q2    | \_\_idiv\_\_(self, Q2)      |
| Divisão Inteira          | Q1 //= Q2   | \_\_ifloordiv\_\_(self, Q2) |
| Resto da Divisão         | Q1 %= Q2    | \_\_imod\_\_(self, Q2)      |
| AND                      | Q1 &= Q2    | \_\_iand\_\_(Q1, Q2)        |
| OR                       | Q1 \|= Q2    | \_\_ior\_\_(Q1, Q2)        |
| XOR                      | Q1 ^= Q2    | \_\_ixor\_\_(Q1, Q2)        |
| Deslocar bits a esquerda | Q1 <<= Q2   | \_\_ilshift\_\_(self, bits) |
| Deslocar bits a direita  | Q1 >>= Q2   | \_\_irshift\_\_(self, bits) |

```python
class Quantidade:
    def __init__(self, val):
        self.__valor = val
        
    def __add__(self, outro):
        print("-------------")
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor + outro
        else:
            novo = self.__valor + outro.__valor
        return Quantidade(novo)
    def __radd__(self, outro):
        print("----R----")
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor + outro
        else:
            novo = self.__valor + outro.__valor
        return Quantidade(novo)
    
    def __gt__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            return self.__valor > outro
        else:
            return self.__valor>outro.__valor
    
    def __iadd__(self, outro):
        print("-------------")
        if isinstance(outro, int) or isinstance(outro, float):
            self.__valor += outro
        else:
            self.__valor += outro.__valor
        return Quantidade(self.__valor)
    
    def __str__(self):
        return "Quantidade(" + str(self.__valor) +")"
    
    
Q1 = Quantidade(10)
Q2 = Quantidade(5)

print(Q1+Q2)
# não altera o valor
Q1.__valor=20
print(Q1+20)
print(Q1+20.98)
print(20.98+Q1)
R1=20.98+Q1
print(type(R1))

print(R1>10)

R1+=2
print(R1)
```

```python
class Distancia:
    def __init__(self, val):
        self.__valor = val

    def __add__(self, outro):
         if isinstance(outro, int) or isinstance(outro, float):
             novo = self.__valor + outro
         else:
             novo = self.__valor + outro.__valor
         return Distancia(novo)   

    def __iadd__(self, outro):
         if isinstance(outro, int) or isinstance(outro, float):
             self.__valor += outro
         else:
             self.__valor += outro.__valor
         return Distancia(self.__valor)   

    def __radd__(self, outro):
         if isinstance(outro, int) or isinstance(outro, float):
             novo = self.__valor + outro
         else:
             novo = self.__valor + outro.__valor
         return Distancia(novo)   

    def __sub__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor - outro
        else:
            novo = self.__valor - outro.__valor
        return Distancia(novo)

    def __isub__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            self.__valor -= outro
        else:
            self.__valor -= outro.__valor
        return Distancia(self.__valor)

    def __rsub__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor - outro
        else:
            novo = self.__valor - outro.__valor
        return Distancia(novo)

    def __mul__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor * outro
        else:
            novo = self.__valor * outro.__valor
        return Distancia(novo)

    def __imul__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            self.__valor *= outro
        else:
            self.__valor *= outro.__valor
        return Distancia(self.__valor)

    def __rmul__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor * outro
        else:
            novo = self.__valor * outro.__valor
        return Distancia(novo)

    def __pow__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor ** outro
        else:
            novo = self.__valor ** outro.__valor
        return Distancia(novo)

    def __ipow__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            self.__valor **= outro
        else:
            self.__valor **= outro
        return Distancia(self.__valor)

    def __rpow__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor ** outro
        else:
            novo = self.__valor ** outro.__valor
        return Distancia(novo)

    def __truediv__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor / outro
        else:
            novo = self.__valor / outro.__valor
        return Distancia(novo)

    def __itruediv__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            self.__valor /= outro
        else:
            self.__valor /= outro.__valor
        return Distancia(self.__valor)

    def __rtruediv__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor / outro
        else:
            novo = self.__valor / outro.__valor
        return Distancia(novo)

    def __floordiv__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor // outro
        else:
            novo = self.__valor // outro.__valor
        return Distancia(novo)

    def __ifloordiv__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            self.__valor //= outro
        else:
            self.__valor //= outro.__valor
        return Distancia(self.__valor)

    def __rfloordiv__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor // outro
        else:
            novo = self.__valor // outro.__valor
        return Distancia(novo)

    def __mod__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor % outro
        else:
            novo = self.__valor % outro.__valor
        return Distancia(novo)

    def __imod__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            self.__valor %= outro
        else:
            self.__valor %= outro.__valor
        return Distancia(self.__valor)

    def __rmod__(self, outro):
        if isinstance(outro, int) or isinstance(outro, float):
            novo = self.__valor % outro
        else:
            novo = self.__valor % outro.__valor
        return Distancia(novo)

    def __str__(self):
        return "Distancia(" + str(self.__valor) + ")"

D1 = Distancia(10)
D2 = Distancia(20)
print(D1+D2)
print(D1+20)
print(10+D2)
D1+=D2
print(D1)
```

## Propriedades

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        
    def __set_idade(self, valor):
        print("---set---")
        if valor > 0 and valor < 120:
            self.__idade = valor
        else:
            self.__idade = None
            
    def __get_idade(self):
        print("---get---")
        return self.__idade
    
    def __del_idade(self):
        print("---del---")
        del self.__idade        
        
    def __set_nome(self, valor):
        print("---set---")
        if len(valor) < 50:
            self.__nome = valor
        else:
            self.__nome = None
            
    def __get_nome(self):
        print("---get---")
        return self.__nome
    
    def __del_nome(self):
        print("---del---")
        del self.__nome
    
    idade = property(__get_idade, __set_idade, __del_idade)
    nome = property(__get_nome, __set_nome, __del_nome)
        
    def __str__(self):
        return self.__nome + " possui " + str(self.idade) + " anos!"
    
P1 = Pessoa("Heitor", 22)
print(P1)
P1.idade = 24
print(P1)
del P1.nome
del P1.idade    
```

## Exercício Conta Bancária (Continuação 3)

1. Use decoradores na classe `ContaBancaria` para definir o acesso aos atributos. 
    - número da conta
    - nome da pessoa
    - saldo inicial (em reais)
    - tipo de conta (corrente, poupança ou investimento)


```python
class ContaBancaria():

    __cont = 0

    @classmethod
    def contar(cls):
        ContaBancaria.__cont += 1 ##
    
    def __init__(self, numConta, nomePessoa, balancoInicial, tipoConta):
        self.contar()
        self.__numConta = numConta
        self.__nome = nomePessoa
        self.__balancoInicial = balancoInicial
        self.__tipoConta = tipoConta
        print("instancia criada")
        print("Total de contas: " + str(ContaBancaria.__cont))
        print("------------------")

    @property
    def numConta(self):
        return self.__numConta

    @numConta.setter
    def numConta(self, valor):
        if isinstance(valor, int):
            self.__numConta = valor

    @numConta.deleter
    def numConta(self):
        del self.__numConta

    @property
    def nome(self):
        return self.__nome

    @nome.setter
    def nome(self, nome):
        if len(nome) < 50:
            self.__nome = nome

    @nome.deleter
    def nome(self):
        del self.__nome

    @property
    def balancoInicial(self):
        return self.__balancoInicial

    @balancoInicial.setter
    def balancoInicial(self, valor):
        if isinstance(valor, int) or isinstance(valor, float):
            self.__balancoInicial = valor

    @balancoInicial.deleter
    def balancoInicial(self):
        del self.__balancoInicial

    @property
    def tipoConta(self):
        return self.__tipoConta

    @tipoConta.setter
    def tipoConta(self, tipo):
        if isinstance(tipo, str):
            if tipo.lower() == "corrente":
                self.__tipoConta = "Corrente"
            elif tipo.lower() == "poupanca" or tipo.lower() == "poupança":
                self.__tipoConta = "Poupança"
            elif tipo.lower() == "investimento":
                self.__tipoConta = "Investimento"

    @tipoConta.deleter
    def tipoConta(self):
        del self.__tipoConta


    
    def deposito(self, quantidade):
        self.__balancoInicial += quantidade
        print("Deposito realizado")

    def saque(self, quantidade):
        self.__balancoInicial -= quantidade
        print("Saque realizado")
        
    def saldo(self):
        return self.__balancoInicial
    
    def __str__(self):
        return "A conta de " + self.__nome +", de número " + self.__numConta + " e tipo " + self.__tipoConta + " possui " + str(self.__balancoInicial) + " Reais."

class ContaCorrente(ContaBancaria):

    def __init__(self, numConta, nomePessoa, balancoInicial, limite):
        super().__init__(numConta, nomePessoa, balancoInicial, "Conta Corrente")
        self.__limite = limite

    def saque(self, valor):
        if self.saldo() < valor:
            print("Saldo insuficiente")
        elif self.saldo() < self.__limite:
            print("Saque superior ao limite delimitado")
        else:
            super().saque(valor)

class ContaDeposito(ContaBancaria):

    def __init__(self, numConta, nomePessoa, balancoInicial, taxaJuros):
        super().__init__(numConta, nomePessoa, balancoInicial, "Conta Poupança")
        self.__taxaJuros = taxaJuros

    def aplicarJuros(self):
        juros = self.saldo() * (self.__taxaJuros / 100) 
        self.deposito(juros)
        print(f"Juros de {self.__taxaJuros}% aplicado. Seu saldo é: {self.saldo()}")

class ContaInvestimento(ContaBancaria):
    def __init__(self, numConta, nomePessoa, balancoInicial, tipoInvestimento):
        super().__init__(numConta, nomePessoa, balancoInicial, "Investimento")
        self.__tipoInvestimento = tipoInvestimento

    def __str__(self):
        return super().__str__() + f" Tipo de investimento: {self.__tipoInvestimento}"

conta1 = ContaCorrente("1234", "João", 5000, 1000)
print(conta1)
conta1.saque(2000)
conta1.saque(500)

conta2 = ContaDeposito("5678", "Maria", 3000, 5)
print(conta2)
conta2.aplicarJuros()

conta3 = ContaInvestimento("91011", "Pedro", 15000, "Alto Risco")
print(conta3)
```
